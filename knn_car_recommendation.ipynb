## Knn for car recommendation

```python
import sklearn
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd
```

```python
cars = pd.read_csv(r'C:\Users\Bharath\mtcars.csv')
#cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
carfeatures = cars.iloc[:,[1, 3, 4, 6]]
## selecting few features for recommendation 
carfeatures
```

```python
## k = 5 i.e., Number of Neighbors = 5
num_recommendation = 5
k_neighbors = NearestNeighbors(n_neighbors=num_recommendation).fit(carfeatures)
```

```python
# required_features is select the cars based on the features 
required_features = [18, 315, 110, 1.5]
CarsRecommonded = k_neighbors.kneighbors([required_features])
CarsRecommonded = CarsRecommonded[1][0]
CarsRecommonded =  CarsRecommonded.tolist()
print('Following cars are Recommonded ', CarsRecommonded)
```

```python
print('Recommonded Cars details')
templist = []
for num in CarsRecommonded:
    print(cars.iloc[[num],:])
```
